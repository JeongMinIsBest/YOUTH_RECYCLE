```python
import folium
import json

# 파일 경로
geojson_path = '/content/hangjeongdong_전라북도 (1).geojson'

# GeoJSON 데이터 불러오기
with open(geojson_path, 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)
```

```python
# 전주시 데이터만 필터링
jeonju_dong = {
    "type": "FeatureCollection",
    "features": []
}

# 전주시의 행정동 데이터 추출 ('adm_nm' 속성을 사용)
for feature in geojson_data['features']:
    if '전주시' in feature['properties']['adm_nm']:
        jeonju_dong['features'].append(feature)
```

```python
# 전주시 중심 좌표 (대략적인 좌표)
center = [35.824223, 127.147953]

# 전주시 지도를 생성
m = folium.Map(location=center, zoom_start=12)

# GeoJSON 데이터를 지도에 추가
folium.GeoJson(jeonju_dong, name="전주시 행정동").add_to(m)

m
```

```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

# API 호출 URL
url = 'http://openapi.jeonju.go.kr/rest/recyclingbin/getRecyclingBin'

# 발급받은 인증키 입력 (올바르게 URL 인코딩된 인증키)
service_key = 'jJ3LYIpBdYThfK0eAh0FsQQs41nXQaogyelkcuMI6ynKkTlowbHLlTGRXaGv/K3rmPl9K6ciHBHmUwH5DcqKBA=='

# 파라미터 설정 (필터링 없음)
params = {
    'serviceKey': service_key,   # 발급받은 인증키
    'startPage': 1,              # 시작 페이지
    'pageSize': 100              # 최대 페이지 크기 설정
}

# API 호출
response = requests.get(url, params=params)

# 응답 상태 코드 확인 및 출력
if response.status_code == 200:
    # XML 데이터 파싱
    decoded_content = response.content.decode('utf-8')
    root = ET.fromstring(decoded_content)

    # 분리수거함 정보를 담을 리스트 생성
    recycling_bins = []

    # XML에서 분리수거함 정보 추출
    for item in root.findall('.//list'):
        gu = item.find('gu').text.strip()
        dong = item.find('dong').text.strip()
        ins_location = item.find('insLocation').text.strip()
        ins_type = item.find('insType').text.strip()
        posx = item.find('posx').text.strip()  # 경도
        posy = item.find('posy').text.strip()  # 위도
        base_date = item.find('baseDate').text.strip()

        # 추출한 정보를 리스트에 추가
        recycling_bins.append({
            '구': gu,
            '동': dong,
            '설치장소': ins_location,
            '설치유형': ins_type,
            '경도': posx,
            '위도': posy,
            '기준일자': base_date
        })

    # 리스트를 DataFrame으로 변환
    df = pd.DataFrame(recycling_bins)

    # DataFrame을 HTML 테이블 형식으로 변환
    html_output = df.to_html(index=False, escape=False)
    print(html_output)

else:
    print(f"API 요청 실패. 상태 코드: {response.status_code}")
```
