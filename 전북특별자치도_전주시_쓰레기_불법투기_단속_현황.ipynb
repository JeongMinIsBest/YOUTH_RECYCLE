```python
import pandas as pd
import requests
import time

# Load CSV data
df = pd.read_csv('/content/filtered_violation_data.csv')

# Filter rows with '길', '대로', '로', or '동'
df = df[df['위반장소'].str.contains(r'(길|대로|로|동)', na=False)]

# Kakao API Key
api_key = '00791462b1922e859b32fb0ce82d9ceb'  # Replace with your Kakao API key

# Caching to store processed addresses
cache = {}

# Function to get location data
def get_location_info(road_name, api_key, retries=3):
    # Check cache first
    if road_name in cache:
        return cache[road_name]

    url = f"https://dapi.kakao.com/v2/local/search/address.json?query={road_name}"
    headers = {"Authorization": f"KakaoAK {api_key}"}

    for attempt in range(retries):
        try:
            response = requests.get(url, headers=headers, timeout=5)
            if response.status_code == 200:
                try:
                    address_info = response.json()['documents'][0]
                    region = address_info['address']['region_3depth_name']
                    latitude = address_info['y']
                    longitude = address_info['x']
                    # Cache result
                    cache[road_name] = (region, latitude, longitude)
                    return region, latitude, longitude
                except (IndexError, KeyError):
                    return None, None, None
            else:
                print(f"Error {response.status_code} for {road_name}")
                return None, None, None
        except requests.RequestException as e:
            print(f"Attempt {attempt + 1} failed for {road_name}. Error: {e}")
            time.sleep(1)
    return None, None, None

# List to store the results
results = []

# Process each road name
for road_name in df['위반장소']:
    district, latitude, longitude = get_location_info(road_name, api_key)
    results.append([road_name, district, latitude, longitude])
    time.sleep(1)  # Adjust sleep duration as needed

# Save to CSV
locations_df = pd.DataFrame(results, columns=['위반장소', '행정동', 'Latitude', 'Longitude'])
locations_df.to_csv('/content/violation_locations_with_coordinates.csv', index=False, encoding='utf-8-sig')

print("Data with administrative districts and coordinates has been saved.")
```

```python
# Importing the required libraries
import folium
import json
import pandas as pd

# GeoJSON file path and uploaded CSV file path
geojson_path = '/content/hangjeongdong_전라북도.geojson'
csv_path = '/content/violation_locations_with_coordinates1.csv'  # Update this path

# Load the GeoJSON data
with open(geojson_path, 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)

# Filter for Jeonju city administrative districts only
jeonju_dong = {
    "type": "FeatureCollection",
    "features": []
}
for feature in geojson_data['features']:
    if '전주시' in feature['properties']['adm_nm']:
        jeonju_dong['features'].append(feature)

# Define the center of Jeonju (approximate coordinates)
center = [35.824223, 127.147953]

# Create a map centered on Jeonju
m = folium.Map(location=center, zoom_start=12)

# Add GeoJSON data to the map
folium.GeoJson(jeonju_dong, name="전주시 행정동").add_to(m)

# Load the CSV data with violation locations
df = pd.read_csv(csv_path)

# Plot each location from the CSV file as a red dot based on latitude and longitude
for idx, row in df.iterrows():
    lat, lon = row['Latitude'], row['Longitude']
    # Check for null values
    if pd.notna(lat) and pd.notna(lon):
        folium.CircleMarker(
            location=[lat, lon],
            radius=5,               # Size of the dot
            color='red',            # Border color
            fill=True,
            fill_color='red',       # Fill color
            fill_opacity=0.7        # Transparency of the dot
        ).add_to(m)

# Display the map with red dot markers and boundaries
m
```
